## Задание 2

```python
df= pd.read_csv ('Mountains.csv', delimiter = ',')
###Заменяем название колонки, чтобы испльзовать ее ниже в синтаксисе
df.rename( columns = { 'Parent mountain': 'PM' }, inplace=True )
## Создаем DF, содержащий столбец count, чтобы дальше по нему фильтровать
df_pm = df.groupby('PM').PM.count().reset_index(name="count2")

df_pm[df_pm.count2 > 3].head()
```
