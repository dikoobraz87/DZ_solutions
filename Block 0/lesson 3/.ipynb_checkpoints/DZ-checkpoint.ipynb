# Решение

```python
import pandas as pd
import numpy as np

import scipy.stats as st
import matplotlib.pyplot as plt
```

## Задание 1

```python
df = pd.read_csv ('hw_25000.csv', names = ['index', 'height_inches','weight_pound'], header = 0)
```

```python
df['height'] = df['height_inches'] * 2.54
df['weight'] = df['weight_pound'] *0.45

### создаем выборку
sample = df[ (df['height']>= 170) & (df['height'] <= 180)].weight

###Находим средние значения
sample_mean = np.mean (sample)
grid_mean = np.mean(df['weight'])

    
    
## Средний вес выборки и исходных данных    
print (sample_mean)   
print (grid_mean)

```

## проверим на нормальность оба распределения

```python
st.probplot (sample, dist = 'norm',plot=plt)
```

```python
st.probplot (df['weight'], dist = 'norm',plot=plt)
```

```python
st.ttest_1samp (sample,df['weight'].mean())
```

# Вывод: так как P<0.05 - то разница веса получилась статистически значима. Мы отклоняем 0 гипотезу
