## Matplotlib

```python

import matplotlib.pyplot as plt
```

```python
### берем два датасета NBA. Возьмем только данные с 90го года
df_stats = pd.read_csv ('seasons_stats.csv', delimiter= ',')
df_players = pd.read_csv ('players.csv', delimiter= ',')

df_players.drop(df_players.columns[[0]], axis=1, inplace=True)
df_players.index.name = 'Id'
df_stats = df_stats[df_stats['Year'] > 1990]
```

```python
### приводим типы к строкам
df_stats.loc[:,'Age']= df_stats['Age'].astype(int).astype(str)
df_stats.loc[:,'Year']= df_stats['Year'].astype(int).astype(str)
##df_players.head()
##df_stats['Pos'].value_counts()
df_stats.head()
```

## Seaborn

### 2 Посмотрим как распределились набранные по годам  очки по отношению к максимальному по очкам году

```python
import seaborn as sns
```

```python


### находим самый многоочковый сезон
max_points = df_stats[['Year','PTS']].groupby('Year').sum().max()

###Создаем выборку
subset = df_stats[['Year','PTS']].groupby('Year', as_index=False).sum()
subset['maxPTS'] =  float(max_points)
subset
subset['Year'] = subset['Year'].astype(str)



# строим два бара, с максимальным значением и значениям по годам. Не нашел как отсортировать по убвыанию по очкам
f, ax = plt.subplots(figsize=(20, 15))

sns.set_color_codes("pastel")
sns.barplot(x="maxPTS", y="Year", data=subset,
            label="Max PTS", color="b",orient='h')
sns.set_color_codes("muted")
sns.barplot(x="PTS", y="Year", data=subset,
            label="PTS by Year", color="b", orient = 'h')


# Добавляем легенду
ax.legend(ncol=2, loc="lower right", frameon=True)

```
