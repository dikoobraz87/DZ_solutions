# Решение дз по ансамблям

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import auc, roc_curve,mean_squared_error, explained_variance_score
from sklearn.model_selection import StratifiedShuffleSplit
```

## Загружаем данные

```python
data = pd.read_csv ('train.csv')
```

```python
data
```

```python
data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
d_train, d_test, y_train, y_test = train_test_split(data,data.SalePrice, test_size = 0.3, random_state = 777)
    
```

```python
## Определяем категориальные фичи.
```

```python
cat_feat = list(data.dtypes[data.dtypes == object].index)

#закодируем пропущенные значений строкой, факт пропущенного значения тоже может нести в себе информацию
data[cat_feat] = data[cat_feat].fillna('nan')

#отфильтруем непрерывные признаки
num_feat = [f for f in data if f not in (cat_feat + ['ID', 'SalePrice'])]

cat_nunique = d_train[cat_feat].nunique()
print(cat_nunique)
cat_feat = list(cat_nunique[cat_nunique < 30].index)
```

**Создаем признаки для "деревянных" моделей**

1. Заменяем пропуски на специальное значение -999, чтобы деревья могли их отличить
3. Создаем дамми-переменные для категорий

```python
dummy_train = pd.get_dummies(d_train[cat_feat], columns=cat_feat)
dummy_test = pd.get_dummies(d_test[cat_feat], columns=cat_feat)

dummy_cols = list(set(dummy_train) & set(dummy_test))

dummy_train = dummy_train[dummy_cols]
dummy_test = dummy_test[dummy_cols]


X_train = pd.concat([d_train[num_feat].fillna(-999),
                     dummy_train], axis=1)

X_test = pd.concat([d_test[num_feat].fillna(-999),
                     dummy_test], axis=1)
```

### XGBOOST

```python
import xgboost as xgb
```

```python
params = {
'n_estimators' :100, 'learning_rate': 0.08, 'gamma' : 0, 'subsample':0.75, 
                           'colsample_bytree':1, 'max_depth' :7}

reg_xgb = xgb.XGBRegressor(**params)
reg_xgb.fit(X_train, y_train)
```

```python
y_pred_xgb_test = reg_xgb.predict(X_test)
y_pred_xgb_train = reg_xgb.predict(X_train)
```

```python
print('Train:')
explained_variance_score(y_pred_xgb_train,y_train)
```

```python
print('Test:')
explained_variance_score( y_pred_xgb_test,y_test)
```

### HyperOpt

```python
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials

#функция, которую будем МИНИМИЗИРОВАТЬ
def score(params):
    params['max_depth'] = int(params['max_depth'])
    params['n_jobs'] = -1
    print("Training with params : ", params)
    reg = xgb.XGBRegressor(**params)
    reg.fit(X_train, y_train)
    y_pred_xgb_test = reg.predict(X_test)
    evc = explained_variance_score( y_pred_xgb_test,y_test)
    result = {'loss': 1-evc, 'status': STATUS_OK}
    print('TEST explained_variance_score: {0:.4f}'.format(evc))
    return result



space = {
    'task': 'regression',
    'booster': 'gbtree',
    'objective': 'reg:linear',
    'eta': hp.quniform('eta', 0.01, 1, 0.01),
    'gamma': hp.quniform('gamma', 0, 2, 0.1),
    'min_child_weight': hp.quniform('min_child_weight', 0, 10, 1),
    'max_depth': hp.quniform('max_depth', 1, 10, 1),
    'subsample': hp.quniform('subsample', 0.5, 1, 0.1),
    'colsample_bytree': hp.quniform('colsample_bytree', 0.1, 1, 0.1),
    'silent': 1,
    'n_estimators': 50,
    'learning_rate': 0.03
}


trials = Trials()

best = fmin(score, space, algo=tpe.suggest, trials=trials, max_evals=20)
```

```python
best
```

```python
trials.best_trial
```

### Важность признаков


```python
bst = reg_xgb.get_booster()

for kind in ['weight', 'gain', 'cover']:
    imp = pd.Series(bst.get_score(importance_type=kind))
    plt.figure()
    imp.sort_values(ascending=False).iloc[:10].plot(kind='barh', title=kind)
```
