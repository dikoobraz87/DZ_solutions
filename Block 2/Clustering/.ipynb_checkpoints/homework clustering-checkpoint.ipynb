```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans , DBSCAN
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

```python
df = pd.read_excel ('geo_comment.xlsx')
```

```python
df.head()
```

```python
clusters = KMeans(n_clusters=5)
predict = clusters.fit_predict (df[['x','y']])
```

```python
fig, ax = plt.subplots(figsize = (20,10))
ax.scatter(df.x, df.y, c=predict,  s=60, alpha = 0.2)
```

```python
list_inertia = []
n_clusters = np.arange (2,10)
for cl in n_clusters:
    kmean = KMeans (n_clusters= cl)
    kmean.fit_predict (df[['x','y']])
    list_inertia.append (kmean.inertia_)  
        
```

```python
plt.plot(n_clusters, list_inertia)
```

```python
diff = np.diff(list_inertia)
plt.plot(n_clusters[1:], diff)
```

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(n_clusters[1:-1], diff_r)
```

```python
k_opt = n_clusters[np.argmin(diff_r)+1]
k_opt
```
